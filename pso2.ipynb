```python
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True)

data_set = pd.read_csv('https://github.com/taesaksit/project/raw/main/heart.csv').to_numpy()

def initialize_population(num_particles, num_dimensions):


    positions = np.empty(10 , dtype='object')
    velocities = np.empty(10 , dtype='object')
    
    for i in range(10):
        positions[i] = np.random.random( size=(num_particles, num_dimensions))
        velocities[i] = np.random.random( size=(1, num_dimensions)) 
    
    return positions, velocities

num_particles = 2
num_dimensions = 13

```

```python
def evaluate_fitness(population):
    
    perform = []
    size_data = len(data_set)  
    size_pop = len(population)
    


    for i in range(size_pop):
        pd = population[i]
        count = 0

        for j in range(size_data):
            data = data_set[j, :13]
            data_class = data_set[j, 13:]

            # Clone data
            clone_data = np.tile(data, (2, 1))

            edr = np.argmin(np.sqrt(np.sum((clone_data - pd) ** 2, axis=1)))

            if edr == data_class:
                count += 1

        avg = count / size_data
        perform.append((i, count, avg))
        
    #Sort function
    perform_gbest = np.array(perform)    
    sorts = perform_gbest[np.argsort(-perform_gbest[:, 1])]

    #score
    score_gbest = sorts[0,1]
    score_worst = sorts[-1,1]
    
    #Vector
    idx_gbest = int(sorts[0,0])
    idx_gworst = int(sorts[-1,0])
    v_gbest = population[idx_gbest]
    v_gworst = population[idx_gworst]
    

    
    

    
    return score_gbest , score_worst , v_gbest , v_gworst
```

```python

def swarm(pop, velocity , s_gbest, s_gworst, v_gbest, v_gworst):
    
    result = []
    
    w = 0.8
    c1 = 0.5
    c2 = 0.5
    
    for itera in range(100):
        performance = []
        
        for i in range(len(pop)):
            
            
            #prepare data
            v = velocity[i]
            pi = pop[i]
            clone_v = np.tile(v , (2,1))

            #process
            new_v = w * clone_v + c1 * np.random.random()*(v_gworst - pi) + c2 * np.random.random() * (v_gbest - pi)
            result.append(new_v + pi)



            #Evaluate fitness
            new_s_gbest , new_s_gworst , new_v_gbest, new_v_gworst = evaluate_fitness(np.array(result))

            #Condition
            if new_s_gbest > s_gbest:

                #score gbest
                s_gbest = new_s_gbest
                #vector in population
                pop[i] = new_v_gbest
                #vector in parmeter
                v_gbest = new_s_gbest


            elif  s_gworst > new_s_gbest:

                #score gworst
                s_gworst = new_s_gworst
                #vector in population
                pop[i] = new_v_gworst
                #vector in parmeter
                v_gworst = new_v_gworst
            else:
                continue
        print(s_gbest)
          

        
    return
```

```python
# Main

population, velocities = initialize_population(num_particles, num_dimensions)
s_gbest , s_gworst , v_gbest, v_gworst = evaluate_fitness(population)

swarm(population , velocities, s_gbest, s_gworst, v_gbest, v_gworst)


```
